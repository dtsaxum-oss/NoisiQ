## Background: Visualizing Quantum Error Propagation

For NoisiQ's propagation animation, we extend the standard qubit circuit diagram with a visual "error marker" — a colored indicator that sits on the qubit wire(s) carrying an active Pauli error at each step. As the animation advances:

- The circuit diagram is drawn statically in the background
- The current gate being evaluated is highlighted
- The active error marker moves and transforms according to the conjugation rules
  computed by the Week 2 Pauli-frame engine
- A text panel below the diagram shows the current Pauli frame state in notation
  (e.g., "Qubit 0: X, Qubit 1: I")

**Design principles for the visualization:**
- The visuals must accurately reflect the Pauli frame state at every step. 
- The diagram should be readable at a glance. We will seek to avoid visual clutter. 
- The animation renderer must gracefully handle circuits up to 16 qubits and depth 64, an amendable amount of qubits and depth to prevent render errors.

## Physics Requirements for Week 3

**Animation must be a faithful representation of the Pauli frame.**
At every frame of the animation, the error markers displayed must exactly match the Pauli frame state computed by backend. The visualization layer only displays the computed state.

**Error identity must be visually distinguishable.**
X, Y, and Z errors must be rendered in visually distinct ways (color, label, or both) so a user can immediately identify which type of error is present on which qubit. 

**Multi-qubit error states must be shown simultaneously.**
If an error has spread to multiple qubits (e.g., after a CNOT), all affected qubits must show their error markers at the same time.

```python
# 1. Draw a static Bell state circuit
import noisiq as nq

circuit = nq.Circuit(num_qubits=2)
circuit.add_gate(nq.gates.H, qubits=[0])
circuit.add_gate(nq.gates.CNOT, qubits=[0, 1])

nq.visualize.draw(circuit)

# 2. Inject an X error and run the propagation animation
nq.visualize.propagation(
    circuit,
    error=nq.noise.PauliError(qubit=0, operator="X"),
    inject_at=0,
    export_gif=True,
    gif_path="week3_bell_propagation.gif"
)

# 3. Launch the interactive panel
nq.visualize.interactive(circuit)
```
