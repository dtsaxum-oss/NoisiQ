## Background: Visualizing Quantum Error Propagation

For NoisiQ's propagation animation, we extend the standard qubit circuit diagram with a visual "error marker" — a colored indicator that sits on the qubit wire(s) carrying an active Pauli error at each step. As the animation advances:

- The circuit diagram is drawn statically in the background
- The current gate being evaluated is highlighted
- The active error marker moves and transforms according to the conjugation rules
  computed by the Week 2 Pauli-frame engine
- A text panel below the diagram shows the current Pauli frame state in notation
  (e.g., "Qubit 0: X, Qubit 1: I")

**Design principles for the visualization:**
- The visuals must accurately reflect the Pauli frame state at every step. 
- The diagram should be readable at a glance. We will seek to avoid visual clutter. 
- The animation renderer must gracefully handle circuits up to 16 qubits and depth 64, an amendable amount of qubits and depth to prevent render errors.

## Physics Requirements for Week 3

**Animation must be a faithful representation of the Pauli frame.**
At every frame of the animation, the error markers displayed must exactly match the Pauli frame state computed by backend. The visualization layer only displays the computed state.

**Error identity must be visually distinguishable.**
X, Y, and Z errors must be rendered in visually distinct ways (color, label, or both) so a user can immediately identify which type of error is present on which qubit. 

**Multi-qubit error states must be shown simultaneously.**
If an error has spread to multiple qubits (e.g., after a CNOT), all affected qubits must show their error markers at the same time.

This notebook demonstrates the core capabilities of the NoisiQ MVP:
1. **High-Performance Backend**: Using `Stim` for polynomial-time Clifford simulation.
2. **Layer-by-Layer Tracking**: Explicitly sampling and recording where errors originate.
3. **Interactive Visualization**: Stepping through error propagation with `ipywidgets` and `matplotlib` (Qiskit no longer required for visualization).
4. **Pauli Error Propagation**: Visualizing gate-level error propagation with X, Y, Z labels on wires.

````python
import numpy as np
import matplotlib.pyplot as plt
from noisiq.ir import Circuit, gates
from noisiq.noise import depolarizing_error, bit_flip_error
from noisiq.visualization import Visualizer
````

## 1. Building a Circuit

We'll start by building a simple Bell state circuit ($|00\rangle + |11\rangle$).

````python
c = Circuit(n_qubits=2, name="Bell State")
c.add_gate(gates.H, (0,))
c.add_gate(gates.CNOT, (0, 1))

print(c)
````

## 2. Configuring Noise

We can attach noise models to specific gates in the circuit. Here, we'll add a 10% depolarizing error to the CNOT gate (index 1).

````python
# Noise config maps gate_index to a noise model
noise_config = {
    1: depolarizing_error(0.1)
}

print(f"Noise at CNOT: {noise_config[1]}")
````

## 3. Interactive Visualization

The `Visualizer` class runs the simulation and provides an interactive slider to step through the execution. 

**Note**: In the visualization, gates that trigger a stochastic error are highlighted, and Pauli errors (X, Y, Z) are displayed directly on the wires as they propagate.

````python
vis = Visualizer(c)
vis.simulate(noise_config=noise_config, seed=42)
vis.show()
````

## 4. Exporting Animations

We can also export the error propagation as a GIF animation to share or include in reports.

````python
# Export the propagation animation
vis.export_animation('propagation.gif')
print("Animation saved to propagation.gif")
````

## 5. Complex Example: 5-Qubit GHZ State

Let's see how errors propagate in a larger circuit.

````python
n = 5
ghz = Circuit(n_qubits=n, name="5-Qubit GHZ")
ghz.add_gate(gates.H, (0,))
for i in range(n - 1):
    ghz.add_gate(gates.CNOT, (i, i + 1))

# Add bit-flip noise to the first few CNOTs
ghz_noise = {
    1: bit_flip_error(0.5),
    2: bit_flip_error(0.5)
}

vis_ghz = Visualizer(ghz)
vis_ghz.simulate(noise_config=ghz_noise, seed=123)
vis_ghz.show()
````
