## Physics Requirements for Week 1

Every gate added to a circuit must satisfy the following physical requirement:

**Unitarity:** A gate matrix U must satisfy U†U = I

Non-unitary matrices represent non-physical operations and must be rejected
at the point of gate construction.

This check will be implemented in `gates.py` and tested in `tests/test_gates.py`.

## Week 1 Demo

*This section will be filled in once the Python files above are complete.
It will demonstrate:*

```python
# 1. Importing the package
import noisiq as nq

# 2. Add gates
# 3. Verify a gate is unitary
# 4. Build a Bell state circuit (H on qubit 0, CNOT on qubits 0-1)
# 5. Print a circuit summary
# 6. Confirm the circuit rejects invalid inputs
```

````python
import noisiq as nq
import numpy as np

# 1. Importing the package works
print(f"Successfully imported noisiq version {nq.__version__}\n")

# 2. Accessing standard gates
print("--- Accessing Standard Gates ---")
print(f"Hadamard gate: {nq.ir.H}")
print(f"CNOT gate: {nq.ir.CNOT}")

# 3. Verifying unitarity check
print("\n--- Verifying Unitarity Check ---")
print("Attempting to create a non-unitary gate...")
try:
    non_unitary_matrix = np.array([[1, 2], [3, 4]])
    nq.ir.Gate(name="NonUnitary", matrix=non_unitary_matrix, num_qubits=1)
except ValueError as e:
    print(f"Successfully caught expected error: {e}")

# 4. Building a Bell state circuit
print("\n--- Building a Bell State Circuit ---")
bell_circuit = nq.Circuit(n_qubits=2, name="bell_state")
bell_circuit.add_gate(nq.ir.H, qubits=[0])
bell_circuit.add_gate(nq.ir.CNOT, qubits=[0, 1])
print("Circuit built successfully.")

# 5. Printing the circuit summary
print("\n--- Circuit Summary ---")
print(bell_circuit)

# 6. Confirming the circuit rejects invalid inputs
print("\n--- Verifying Circuit Input Validation ---")
# --- Test 1: Qubit index out of bounds ---
try:
    bell_circuit.add_gate(nq.ir.X, qubits=[2])
except ValueError as e:
    print(f"Successfully caught expected error (qubit out of bounds): {e}")

# --- Test 2: Wrong number of qubits for a gate ---
try:
    bell_circuit.add_gate(nq.ir.H, qubits=[0, 1])
except ValueError as e:
    print(f"Successfully caught expected error (wrong qubit count): {e}")
````
