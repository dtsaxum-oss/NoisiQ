## Background: Pauli Frame

A Pauli frame acts as a method for tracking errors in a quantum circuit without simulating the full quantum state. Instead of maintaining an exponentially large statevector, the Pauli frame records which Pauli operator (I, X, Y, or Z) is "riding along" on each qubit as an error.

When a gate is applied, the error does not stay fixed, it transforms according to the conjugation rules of that gate. For example, an X error on a qubit passing through a Hadamard gate becomes a Z error on the other side. These are known conjugation rules that are exact and algebraic. Which makes Pauli-frame propagation fast, deterministic, and verifiable.

- It is computationally cheap, requiring no matrix exponentiation and no statevectors
- The conjugation rules are well-established and testable against known identities
- It forms the foundation for the probabilistic noise channels for later development
- It scales efficiently to larger circuits, which is critical for our large-circuit mode

**Key conjugation identities:**
```
H X H† = Z
H Z H† = X
CNOT (XI) CNOT† = XX
CNOT (IX) CNOT† = IX
CNOT (ZI) CNOT† = ZI
CNOT (IZ) CNOT† = ZZ
```
